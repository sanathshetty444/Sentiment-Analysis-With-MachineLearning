```python
import turicreate as tc
```

```python
products=tc.SFrame('amazon_baby.sframe')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
for each in selected_words:
    products[each]=products['word_count'].apply(lambda x : x[each] if each in x.keys() else 0)
    print(products[each].sum, each)
```

```python
products
```

```python
diaper_champ_reviews = products[products['name'] == 'Baby Trend Diaper Champ']
```

```python
len(diaper_champ_reviews)
```

```python
products = products[products['rating'] != 3]
```

```python
products['sentiment'] = products['rating'] >= 4
```

```python
products.head(20)
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
sentiment_model = tc.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=selected_words,
                                                     validation_set=test_data)
```

```python
sentiment_model
```

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
```

```python
diaper_champ_reviews['predicted_sentiment'] = sentiment_model.predict(diaper_champ_reviews, output_type='probability')
```
